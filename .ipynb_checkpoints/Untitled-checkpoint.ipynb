```python
from selenium import webdriver  # # 驱动浏览器
from selenium.webdriver.common.by import By  #选择器
from selenium.webdriver.common.keys import Keys   #按键
from selenium.webdriver.support.wait import WebDriverWait  #等待页面加载完毕，寻找某些元素
from selenium.webdriver.support import expected_conditions as EC  ##等待指定标签加载完毕
import time
```

```python
browser = webdriver.Chrome()  
browser.get("https://search.jd.com/Search?keyword=%E7%83%AD%E6%B0%B4%E5%99%A8&enc=utf-8&wq=%E7%83%AD%E6%B0%B4%E5%99%A8&pvid=9b85203521db4369805cd544ee0d298d")

def na():
    price = browser.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div[2]/strong/i')
    store = browser.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div[3]/a/em')
    pinjia = browser.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div[4]/strong')
    s = browser.find_elements(By.XPATH, '//*[@id="J_goodsList"]/ul/li/div/div[5]/span/a')

    for i, j, m, n in zip(price, store, pinjia, s):
        print(i.text)
        print(j.text)
        print(m.text)
        print(n.text)
        print("==========")
        
na()
for i in range(50):
    browser.find_element(By.XPATH, '//*[@id="J_bottomPage"]/span[1]/a[9]').click()
    time.sleep(3)
    print(f"=======第{i + 2}页正在下载=====")
    na()
```
