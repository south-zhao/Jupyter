```python
import turtle
def restore(p):
    """
    将画笔恢复到原来的位置p点
    """
    turtle.penup()
    turtle.setpos(p[0], p[1])  # 将画笔移动到起点
    turtle.pendown()
    turtle.seth(p[2])  # 设置画笔的方向

def get_point():
    """
    获得当前位置
    """
    x, y = turtle.pos()  # 获取位置
    d = turtle.heading()  # 获取方向
    return x, y, d  # 返回当前状态

def Generator(A, B, L, n):
    """
    生成器函数
    A---起始点
    B---结束点
    L---线段的长度
    n---递归的次数
    """
    if n == 1:
        restore(A)
        turtle.forward(L/3)
        turtle.left(60)
        turtle.forward(L/3)
        turtle.right(120)
        turtle.forward(L/3)
        turtle.left(60)
        turtle.setpos(B[0], B[1])
    else:
        restore(A)
        turtle.pencolor((59, 209, 207))
        turtle.forward(L/3)
        turtle.left(60)
        C = get_point()
        turtle.forward(L/3)
        turtle.right(120)
        D = get_point()
        turtle.forward(L/3)
        turtle.left(60)
        E = get_point()
        turtle.pencolor('white')

        Generator(A, C, L/3, n-1)
        Generator(C, D, L/3, n-1)
        Generator(D, E, L/3, n-1)
        Generator(E, B, L/3, n-1)
    return True

if __name__ == '__main__':
    turtle.hideturtle()
    turtle.colormode(255)
    turtle.bgcolor((59, 209, 207))
    turtle.pencolor('white')
    turtle.pensize(2)

    A = (-450, 0, 0)
    B = (450, 0, 0)
    L = 900
    n = 6

    restore(A)
    Generator(A, B, L, n)
```
